```python
from math import pi, sin

x = []
dx = pi/100
fx = []
for i in range(0,101):
    x.append(i*dx)

for j in x:
    fx.append(sin(j))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x,fx,'r-');
plt.xlabel('x');
plt.ylabel('sin(x)');
plt.grid();
plt.fill(x,fx,'lightblue');
```

```python
I = 0
for i in range(0, len(fx)):
        I += fx[i]*dx
```
